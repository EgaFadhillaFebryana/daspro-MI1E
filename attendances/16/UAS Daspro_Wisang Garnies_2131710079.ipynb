```java
import java.util.Scanner;
Scanner input = new Scanner(System.in);

// Deklarasi Array
String namaa;
String [] namaBarang = new String[8];
int [] dataBarang = new int [8];
int [] dataMasuk = new int [8];
int [] dataKeluar = new int [8];
int [] dataRusak = new int [8];


// Fungsi untuk menginputkan data barang
static void inputDataBarang(){
    for(int i = 0 ; i < dataBarang.length;i++){
        System.out.print("Masukan Nama Barang   : ");
        String nama = input.next();
        namaBarang[i] = nama;
        System.out.print("Masukan Jumlah Barang : ");
        int jum = input.nextInt();
        dataBarang[i] = jum;
        System.out.print("Apakah Anda Ingin Menginputkan Barang Lagi ? (Ya/Tidak)");
        char konfirm = input.next().charAt(0);
        if (konfirm == 'y' || konfirm == 'Y' ) continue;
        else break;
    }
}

// Fungsi untuk menginputkan barang masuk
static void inputBarangMasuk(){
    for(int i = 0 ; i<namaBarang.length;i++){
        if(namaBarang[i] == null) continue;
        System.out.println((i+1)+". "+namaBarang[i]+" : "+dataBarang[i]);
    }
    System.out.print("Masukan Angka Barang         : ");
    int angka = input.nextInt();
    System.out.print("Masukan Jumlah Barang Masuk  : ");
    int jum = input.nextInt();
    dataMasuk[angka - 1] = jum;
    dataBarang[angka - 1] += jum;
}

// Fungsi untuk Pengeluaran barang
static void inputBarangKeluar(){
    for(int i = 0 ; i<namaBarang.length;i++){
        if(namaBarang[i] == null) continue;
        System.out.println((i+1)+". "+namaBarang[i]+" : "+dataBarang[i]);
    }
    System.out.print("Masukan Angka Barang          : ");
    int angka = input.nextInt();
    System.out.print("Masukan Jumlah Barang Keluar  : ");
    int jum = input.nextInt();
    dataKeluar[angka - 1] = jum;
    dataBarang[angka - 1] -= jum;
}

// Fungsi untuk mendata barang rusak
static void inputBarangRusak(){
    for(int i = 0 ; i<namaBarang.length;i++){
        if(namaBarang[i] == null) continue;
        System.out.println((i+1)+". "+namaBarang[i]+" : "+dataBarang[i]);
    }
    System.out.print("Masukan Angka Barang          : ");
    int angka = input.nextInt();
    System.out.print("Masukan Jumlah Barang Rusak   : ");
    int jum = input.nextInt();
    dataRusak[angka - 1] = jum;
    dataBarang[angka - 1] -= jum;
}

// Fungsi untuk Melaporkan data barang
static void laporanDataBarang(){
    for(int i = 0 ; i<namaBarang.length;i++){
        if(namaBarang[i] == null) continue;
        System.out.println("Nama Barang    : " + namaBarang[i]);
        System.out.println("Jumlah Barang  : " + dataBarang[i]);
        System.out.println("Barang Masuk   : " + dataMasuk[i]);
        System.out.println("Barang Keluar  : " + dataKeluar[i]);
        System.out.println("Barang Rusak   : " + dataRusak[i]);
    }
}

//Fungsi untuk mencari data barang
static void pencarianDataBarang(){
    System.out.print("Masukan Nama Barang          : ");
    namaa = input.next();
    System.out.println(namaa);
    for(int i = 0 ; i < namaBarang.length ; i++){
        if(namaBarang[i] == null){
            continue;
        }
        if(namaBarang[i].equals(namaa)){
            System.out.println("Nama Barang          : " + namaBarang[i]);
            System.out.println("Jumlah Barang        : " + dataBarang[i]);
            System.out.println("Jumlah Barang Masuk  : " + dataMasuk[i]);
            System.out.println("Jumlah Barang Keluar : " + dataKeluar[i]);
            System.out.println("Jumlah Barang Rusak  : " + dataRusak[i]);
        }  
    }
}

// Untuk Menampilkan Menu fitur 
while(true){
System.out.println("<===========SELAMAT DATANG==========>");
System.out.println("#===================================#");
System.out.println("|             Menu                  |");
System.out.println("|1. Input Data Barang               |");
System.out.println("|2. Input Barang Masuk              |");
System.out.println("|3. Input Barang Keluar             |");
System.out.println("|4. Input barang Rusak              |");
System.out.println("|5. Pencarian Data Barang           |");
System.out.println("|6. Laporan Data Barang Saat Ini    |");
System.out.println("#===================================#");
System.out.println("                                     ");
System.out.print("▶ Masukkan Angka Menurut Menu = ");
int angka = input.nextInt();

// Menjalankan Pemilihan Menu    
if (angka == 1){
    inputDataBarang();
} else if (angka == 2){
    inputBarangMasuk();
} else if (angka == 3){
    inputBarangKeluar();
} else if (angka == 4){
    inputBarangRusak();
} else if (angka == 5){
    pencarianDataBarang();
} else if (angka == 6){
    laporanDataBarang();
} else {
    System.out.print("(X) Angka yang Anda masukkan tidak valid !!!");
}

// Pengecekkan Menu Kembali    
    System.out.print("Apakah Anda Ingin Kembali Ke Menu ? (Ya/Tidak)");
    char konfirm = input.next().charAt(0);
    if (konfirm == 'y' || konfirm == 'Y' ) continue;
    else break;
}
```
